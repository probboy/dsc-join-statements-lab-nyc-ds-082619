## Display the names of all the employees in Boston.

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
#Your code here
c.execute("""select count(employeenumber) as num_employees, officecode
             from employees join offices using(officecode) group by officecode
             order by num_employees""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
c.execute("""select productname, firstname, lastname
             from employees e join customers c on e.employeenumber = c.salesrepemployeenumber
             join orders using(customernumber)
             join orderdetails using(ordernumber)
             join products using(productcode)""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
c.execute("""select count(productname) as num_sold, firstname, lastname
             from employees e join customers c on e.employeenumber = c.salesrepemployeenumber
             join orders using(customernumber)
             join orderdetails using(ordernumber)
             join products using(productcode)
             group by firstname, lastname""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```
